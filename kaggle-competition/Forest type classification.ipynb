```python
def evaluate_confusion_matrix(cf_matrix,ytest,ypred):
    scikitplot.metrics.plot_confusion_matrix(ytest,ypred)
    rows, columns = cf_matrix.shape
    t = 0
    f = 0
    for i in range(rows):
        for j in range(columns):
            if i == j:
                t = t + cf_matrix[i][j]
            else:          
                f = f + cf_matrix[i][j]
    total = t + f      
            
    print("number of correct predection: " + str(t)+ "/" +str(total)) 
    print("number of incorrect prediction are: " + str(f) + "/" + str(total))
           
```

```python
evaluate_confusion_matrix(cf_matrix1,y_test,y_pred)
```

## Applying Neural Network 

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_accuracy,categorical_crossentropy
```

```python
from keras.layers import Dropout
from keras.optimizers import SGD
```

```python
X_train_plus.shape
```

```python
bad_features = ['Hillshade_9am','Hillshade_3pm','Wilderness_Area4','Soil_Type12','Soil_Type14','Soil_Type15','Soil_Type16','Soil_Type17', 'Soil_Type21','Soil_Type26','Soil_Type29','Soil_Type39']
best_list_of_features2 = list(set(X_train.columns) - set(bad_features))
bad_features2 = ['Soil_Type7','Soil_Type18','Soil_Type9','Soil_Type8','Soil_Type37','Soil_Type19'] 
best_list_of_feature3 = list(set(best_list_of_features2)-set(bad_features2))
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(units=27,activation='relu',input_dim=X_train_plus[best_list_of_feature3].shape[1]))
    model.add(Dropout(0.2))
    model.add(Dense(units=500,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(units=250,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(units=125,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(units=7,activation='softmax'))
    model.add(Dropout(0.2))
    optimizer = SGD(lr=0.01, decay=1e-6, momentum=0.99, nesterov=True)
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
model = create_model()
```

```python
model.fit(X_train_plus[best_list_of_feature3],y_train_plus,epochs=100,validation_split=0.1,batch_size=10)
```

```python
y_pred = model.predict(X_test_plus[best_list_of_feature3])
```

```python
y_pred = model.predict_classes(X_test_plus[best_list_of_feature3])
for i in range(len(y_pred)):
    y_pred[i] = y_pred[i] + 1
```

```python
cf_matrix = confusion_matrix(y_test,y_pred)
evaluate_confusion_matrix(cf_matrix,y_test,y_pred)
```
