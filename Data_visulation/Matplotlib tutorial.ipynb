## Simple ploting in matplotlib



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x,'g-',label='linear')
plt.plot(x,x**2,c='r',label='qudratic')
plt.plot(x,x**3,c='b',label='cubic')

plt.legend(loc=1)

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Title")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,8,16],'ro')
plt.ylabel("squared numbers")
plt.axis([0,5,0,20])
plt.show()
```

```python
#ploting various lines togather
x = np.arange(0.,5.,0.2)

plt.plot(x,x,'r-',x,x**2,'b^')
plt.show()
```

## Ploting with keyword strings

```python
data = {'a':np.arange(50),
        'c':np.random.randint(0,50,50),
         'd':np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100 

# print(data)
plt.scatter('a','b',c='c',s='d',data=data)
plt.xlabel("a")
plt.ylabel("y")
plt.title("ramdom fun")
plt.show()
```

## working with multiple figure and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## stack plot

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','b'])

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## How to get data from file

```python
import csv
```

```python
x = list()
y = list()
with open("example.txt",'r') as file:
    plots = csv.reader(file,delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x,y,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x,y = np.loadtxt('example.txt',delimiter=',',unpack=True)
plt.plot(x,y,c='r',label='data from the file',linewidth=2)
plt.grid(True,color='g',linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```

## Getting data from the internet



```python
import matplotlib.pyplot as plt
import numpy as np
import urllib
import matplotlib.dates as mdates
```

```python
def bytespdate2num(fmt,encoding='utf-8'):
    strconverter = mdates.strpdate2num(fmt)
    def bytesconverter(b):
        s = b.decode(encoding)
        return strconverter(s)
    return bytesconverter


def graph_data():
    
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    source_code = urllib.request.urlopen(stock_price_url).read().decode()
    stock_data = list()
    for line in source_code.split("\n"):
        split_line = line.split(',')
        if len(split_line) == 6:
            if 'values' not in line:
                stock_data.append(line)
    
    data,colsep,highp,lowp,openp,volumep = np.loadtxt(stock_data,
                                                     delimiter = ',',
                                                     unpack=True,
                                                      # %Y = full year. 2015
                                                          # %y = partial year 15
                                                          # %m = number month
                                                          # %d = number day
                                                          # %H = hours
                                                          # %M = minutes
                                                          # %S = seconds
                                                          # 12-06-2014
                                                          # %m-%d-%Y

                                                     converters={0: bytespdate2num('%Y-%m-%d')})
    
    plt.xlabel("date")
    plt.ylabel("price")
    plt.legend()
    plt.show()
graph_data()
```
