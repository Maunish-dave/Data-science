### how to write text on top of the barcharts

```python
ax = sns.barplot(x='day',y='tip',data=tips)
```

```python
#ax.patches contains the object which contains the characteristics of each 
#bar chart ax.patches.get_height() and get_width() gives the height and width of the bar 
#charts
#other p funtions are p.get_x()
for p in ax.patches:
    print(p.get_height())
```

For more information on the ax.text see following link (https://matplotlib.org/users/text_intro.html)


```python
sns.set(style='darkgrid')
ax = sns.barplot(x='day',y='tip',data=tips)
total = float(len(tips))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2,height,'{:1.2f}'.format(height),ha='center',color='red')
    
```

to plot the histogram vertically we have a parameter 
called "vertivcal" which have bool value

```python
sns.distplot(label,color='blue',vertical=True)
```

```python
sns.distplot(label,hist=False,color="purple",rug=True)
```

```python
sns.distplot(label,kde=False,color='blue',vertical=False,rug=True,bins=10)
```

data: which contains the data to be ploted

color: color of the bars eg 'red', 'green', 'blue', etc

kde: bool value for displaying kde

# Stripplots

strip plot displays all the data points on the graph


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.stripplot(tips['tip'], color='green')
```

```python
sns.stripplot(x=tips['total_bill'],color='yellow')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='spring')
```

jitter parameter is used to show dispersed datasets

default value is True 

jitter can also have the float value
such as 0.1,0.02 etc

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False) #display all the points in same line
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=0.2)  #more dispersed jitter
```

For displaying horizontal values we need to change x and y

```python
sns.stripplot(x='total_bill',y='day',data=tips,jitter=False) #display all the points in same line
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=tips) #display all the points in same line
```

Stripplot has a new attribute called "linewidth" which will give 
border to the data circles

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=1,linewidth=1) #display border around circle
```

dodge parameter will seperates hue attributes

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True,dodge=True,palette='winter_r')
```

it also contains order parameter

using different shape for the ponit
use marker attribute

```python
sns.stripplot(x='sex',y='tip',data=tips,order=['Female','Male'],jitter=False,palette="spring",marker='D',size=5)
```

using edege color and aplha

```python
sns.stripplot(x='sex',y='tip',data=tips,order=['Female','Male'],jitter=False,marker='D',size=15,edgecolor='grey',alpha=0.3)
```

#### summary of the parameter on stripplot
x , y ,data, order, color, and palette same as others

jitter: to split the datapoints having same value : bool or value

marker: shape of marker to be displayed eg 'D' for diamond

edgecolor: used to display the edge color

aplha: transparency of the data points

size: size of the datapoints



we can plot stripplot and boxplot togather

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,palette='husl')
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,palette='husl')
sns.violinplot(x='day',y='total_bill',data=tips)
```

# pairgrid

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
x= sns.PairGrid(iris)
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist) #in order to display other graphs we need to use
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris,hue='species')
x = x.map_diag(plt.hist) #in order to display other graphs we need to use
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,hue='species',palette='husl')
x = x.map_diag(plt.hist) #in order to display other graphs we need to use
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,hue='species')
x = x.map_diag(plt.hist,histtype='step') #in order to display other graphs we need to use map_offdiag
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,hue='species')
x = x.map_diag(plt.hist,histtype='step',linewidth=4) #in order to display other graphs we need to use map_offdiag
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

vars parameter of PairGrid is used to define the columns to be used from the data

```python
x = sns.PairGrid(iris,vars=['petal_length','petal_width'],hue='species',palette='winter_r')
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris,vars=['petal_length','petal_width'],hue='species',palette='winter_r')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,vars=['petal_length','petal_width'],hue='species',palette='winter_r')
x = x.map_diag(plt.hist,edgecolor='grey')
x = x.map_offdiag(plt.scatter,edgecolor='black')
x = x.add_legend()
```

inorder to get different columns in x axis and y axis we need to 

use x_vars and y_vars instead of vars

```python
x = sns.PairGrid(iris,x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_length'],hue='species')
x  = x.map(plt.scatter)
x = x.add_legend()
```

Fot different plots on the upper triangle and lower triangle
use sns.map_upper and sns.map_lower

```python
x = sns.PairGrid(iris,hue='species')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

to display differnet marker for differnt hue we need to pass a dictionary parameter
hue_kws={}

```python
x = sns.PairGrid(iris,hue='species',hue_kws={'marker':['D','s','+']})
x = x.map(sns.scatterplot, edgecolor='grey')
x = x.add_legend()
```

# Clustermap

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()  
```

we need to add caterogical data like month in this case at the beginning of the data 
for producing heatmap of data

```python
flights = flights.pivot('month','year','passengers')
```

```python
sns.heatmap(flights)
```

according to the heapmap we can see as the year passed number of passenger has increased

and other usual trend has been that number of passenger are maximum during the mont june, july,august, and september
 
and mnumber of passengers in 1960 were atleast above 400.

## clusetermap

clustermap will cluster togather the similar data



```python
sns.clustermap(flights)
```

clearly visible in this cluster map the months june july august and september are clustered togather

and year 1955 to 1956 are cluster togather

If we want to cluster only by rows we have a paramter known as col_cluster which is bool parameter

```python
sns.clustermap(flights,col_cluster=False)
```

similarly for clustering the columns set row_cluster to false

```python
sns.clustermap(flights,row_cluster=False)
```

to change color of the clustermap use cmap parameter instead of paletter 

and to adjust the width between the cluster box we can add linewidth parameter


```python
sns.clustermap(flights,cmap='coolwarm',col_cluster=False,linewidth=1)
```

**standardize and normalize data** 

we can standardize data across the rows and columns 
and normalize our data across rows and columns 
using standardize_scale and z_score 

for rows = 0 for columns = 1

```python
sns.clustermap(flights,cmap='winter_r',standard_scale=0)
```

```python
sns.clustermap(flights,cmap='winter_r',z_score=1)
```

# Kdeplots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mean = [0,0]
cov = [[0.2,0],[0,3]]

x_axis , y_axis = np.random.multivariate_normal(mean,cov,size=40).T
```

```python
sns.kdeplot(x_axis)
```

inorder to give shade under the graph use shade parameter

```python
sns.kdeplot(x_axis,shade=True,color='purple',bw=1.5)
```

two variable kde plot can also be plotted

```python
sns.kdeplot(x_axis,y_axis)
```

```python
sns.kdeplot(x_axis,y_axis,cmap='coolwarm',n_levels=35)
```

```python
sns.kdeplot(x_axis,y_axis,cmap='coolwarm',shade=True) #this wil give shade to two dimensional kde
```

inorder to make kde plot vertical we need to set vertical parameter to True

```python
sns.kdeplot(x_axis,vertical=True)
```

# Regplot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,marker="D",scatter_kws={"color":'blue'},line_kws={"color":"red","linewidth":3.1})
```

inorder to get the mean of the size in tips dataset use x_estimator to mean

```python
tips = sns.load_dataset('tips')
```

```python
sns.regplot(x='size',y='total_bill',data=tips)
```

inorder to spread the data use x_jitter value

```python
sns.regplot(x='size',y='total_bill',data=tips,x_jitter=0.08)
```

```python
sns.regplot(x='size',y='total_bill',data=tips,x_jitter=0.08,x_estimator=np.mean,scatter_kws={"color":"red"})
```

we can also change the confidence interval of the reg line by using ci paremeter

```python
sns.regplot(x='size',y='total_bill',data=tips,x_jitter=0.08,ci=100,line_kws={"color":"red"})
```
